```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/wangyiming2022/Desktop/The Right Technology in the Right Places at NYU SPS   (Responses).csv")
```

```python
df.head()
```

```python
majors = df["What degree program, course or certificate are you currently enrolled in? "]

major_counts = majors.value_counts()

plt.pie(major_counts.values, labels=major_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Majors")
plt.show()
```

```python
gender = df["What is your gender? "]

gender_counts = gender.value_counts()

plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Distribution of gender")
plt.show()
```

```python
location = df["Where do you usually do MOST of your studying? "]

location_counts = location.value_counts()

plt.pie(location_counts.values, labels=location_counts.index, autopct='%1.1f%%')
plt.title("Distribution of location")
plt.show()
```

```python
equipment = df["What computer do you usually use for MOST of your studying? "]

equipment_counts = equipment.value_counts()

plt.pie(equipment_counts.values, labels=equipment_counts.index, autopct='%1.1f%%')
plt.title("equipment")
plt.show()
```

```python
reason = df["The software/technologies used for your study, do you use them on your own computer or does it have to be run on a school lab computer? "]

reason_counts = reason.value_counts()

plt.pie(reason_counts.values, labels=reason_counts.index, autopct='%1.1f%%')
plt.title("reason")
plt.show()
```
